```python
import pandas as pd
import pathlib 
import sys 
import csv

source_folder = 'C:\\Users\\JoseParra\\OneDrive - Red Acre Ltd\\ARCHIVES\\SendGrid\\Kot4x\\csv'
output = 'test_output.csv'

excluded_files = [ 'combined_kot.csv' ]
source_path = pathlib.Path(source_folder)

# Get list of CSV files in the source folder
files_list = list(source_path.glob('*.csv'))
print(f'found {len(files_list)} files to process')

# write to output CSV the names of the files processed, excluding the ones in excluded_files
with open(output, 'w', newline='', encoding='utf-8') as f:
    csv_writer = csv.writer(f)
    csv_writer.writerow([f'-- List of pricesses files --'])
    for item in files_list:
        if item.name in excluded_files:
            print(f'SKIPPING file name: {item.name}')
        else:
            print(f'processing file name: {item.name}')
            csv_writer.writerow([f'--- Start of file: {item.name} ---'])
             
```
